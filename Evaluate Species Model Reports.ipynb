This notebook runs through some tests of the GAP species reports after they have been pulled together into a MongoDB collection, which affords us the ability to run various queries and evaluation routines on the combined information.

```python
from IPython.display import display
from bis2 import mlab
```

```python
bisDB = mlab.getDB("bis")
gapSpeciesCollection = bisDB["gapspecies"]
```

```python
speciesWithNoRefs = []
speciesWithRefErrors = []

for speciesDoc in gapSpeciesCollection.find({"speciesReport":{"$exists":True}},{"identifiersByType":1,"speciesReport.references":1}):
    if isinstance(speciesDoc["speciesReport"]["references"], dict):
        if len(speciesDoc["speciesReport"]["references"].keys()) == 0:
            speciesWithNoRefs.append(speciesDoc["identifiersByType"]["GAP_SpeciesCode"])
        if "ERROR" in speciesDoc["speciesReport"]["references"].values():
            speciesWithRefErrors.append(speciesDoc["identifiersByType"]["GAP_SpeciesCode"])
    elif isinstance(speciesDoc["speciesReport"]["references"], list):
        if len(speciesDoc["speciesReport"]["references"]) == 0:
            speciesWithNoRefs.append(speciesDoc["identifiersByType"]["GAP_SpeciesCode"])
        if "ERROR" in speciesDoc["speciesReport"]["references"]:
            speciesWithRefErrors.append(speciesDoc["identifiersByType"]["GAP_SpeciesCode"])
```

### Species Reports with "ERROR" in references
The following block shows the GAP species codes and references section from the species report JSON where at least one of the references contains an "ERROR" string. These can be either in the newer simple list form of references or those that contain key/value pairs.

```python
for speciesDoc in gapSpeciesCollection.find({"identifiersByType.GAP_SpeciesCode":{"$in":speciesWithRefErrors}},{"identifiersByType":1,"speciesReport.references":1}):
    print (speciesDoc["identifiersByType"]["GAP_SpeciesCode"])
    display (speciesDoc["speciesReport"]["references"])
```

### Species Reports with no references
The following code block displays the list of GAP Species Codes where the species reports contain no references. Is this correct? It seems like all species should have at least some reference.

```python
print (len(speciesWithNoRefs))
display (speciesWithNoRefs)
```

This notebook was built to upload range data files assembled as CSV to the appropriate ScienceBase Items. I ended up doing this off a local Dockerized instance of MongoDB because we ran into performance issues with the GC2-PostgreSQL instance we are trying to use for production. Until we get the necessary capacity to efficiently run these data in ElasticSearch, PostgreSQL/PostGIS, MongoDB, or other technologies we are setting up for the BIS, we may need to deal with things in this way.

This notebook is really just a legacy record of part of what I did so that we don't lose track. It can't really be run anywhere legitimately.

```python
from IPython.display import display
import pysb
```

```python
sb = pysb.SbSession()
username = input("Username: ")
sb.loginc(str(username))
```

```python
rangeItems = sb.find_items("parentId=5951527de4b062508e3b1e79&max=100&fields=identifiers,files")
gapRangeDocs = []
while rangeItems is not None:
    for rangeDoc in rangeItems["items"]:
        gapRangeDocs.append(rangeDoc)
    rangeItems = sb.next(rangeItems)
```

```python
count = 0
for rangeDoc in gapRangeDocs:
    if "files" not in rangeDoc.keys():
        gapSpeciesCode = rangeDoc["identifiers"][next(index for (index,d) in enumerate(rangeDoc["identifiers"]) if d["type"] == "GAP_SpeciesCode")]["key"]
        sbItem = sb.upload_file_to_item({"id":rangeDoc["id"]},"rangeFiles/"+gapSpeciesCode+"_RangeData.csv")
        updateItem["id"] = sbItem["id"]
        updateItem["files"] = sbItem["files"]
        updateItem["files"][0]["title"] = "Range Data Table in CSV format"
        sb.update_item(updateItem)
        count = count + 1
        print (rangeDoc["id"], gapSpeciesCode, count)
```
